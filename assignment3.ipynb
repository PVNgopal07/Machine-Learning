#1. Basic Line Plot

```python
x = np.linspace(-10, 10, 400)
y = x**2

plt.plot(x, y) #used to plot the x,y values
plt.xlabel("x") #lable
plt.ylabel("y = x^2")  #lable
plt.title("Line Plot of y = x^2")
plt.grid(True) #used to present in a grid view
plt.show()
```

#4. Histogram

##a) Simulate 100 random values from a normal distribution using np.random.randn() and plot a histogram.

```python
data = np.random.randn(100)  #random values


plt.hist(data, bins=10, edgecolor='black', color='orange')  # values to specify histogram
plt.xlabel('Value')  #label
plt.ylabel('Frequency')  #label
plt.title('Histogram of 100 Random Values from a Normal Distribution')  #title
plt.show()
```

#5. Pie Chart

##a) Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E'] with values [20, 30, 25, 15, 10].

```python
categories = ['A', 'B', 'C', 'D', 'E']  #categories
values = [20, 30, 25, 15, 10]  #values

plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen', 'gold', 'lightblue'])  # to specify the pie chart
plt.title('Percentage Distribution of Categories') #title
plt.axis('equal') #axis
plt.show()
```

#6. Box Plot (Seaborn)

##a) Create a box plot to visualize the distribution of the marks scored by students in a class: [78, 85, 92, 88, 76, 95, 89, 84, 91, 73].

```python
marks = [78, 85, 92, 88, 76, 95, 89, 84, 91, 73]

plt.figure(figsize=(8, 6))  # adjusting the figure size
sns.boxplot(x=marks)  #creating boxppot
plt.xlabel("Marks")
plt.title("Distribution of Marks Scored by Students")
plt.show()
```

##b) Highlight the median and any outliers.

```python
marks = [78, 85, 92, 88, 76, 95, 89, 84, 91, 73]

plt.figure(figsize=(8, 6))  ## adjusting the figure size
sns.boxplot(x=marks, showmeans=True, medianprops={"color": "red"}, meanprops={"marker":"o", "markerfacecolor":"blue", "markeredgecolor":"black"})  #creating boxplot and highlighting the median
plt.xlabel("Marks")
plt.title("Distribution of Marks Scored by Students")
plt.show()
```

#7. Heatmap (Seaborn)

##a) Create a 5x5 matrix with random integers between 1 and 10 using np.random.randint() and visualize it as a heatmap using Seaborn.

```python
matrix_data = np.random.randint(1, 11, size=(5, 5))

matrix_df = pd.DataFrame(matrix_data)

plt.figure(figsize=(8, 6))  # adjusting the figure size
sns.heatmap(matrix_df, cmap="YlGnBu", linewidths=.5)  #creating heatmap
plt.title("5x5 Matrix Heatmap")
plt.show()
```

##b) Add a color bar and annotate the heatmap with the values.

```python
data = np.random.randint(1, 10, size=(5, 5))
plt.figure(figsize=(8, 6))  # adjusting the figure size
sns.heatmap(data, annot=True, cmap="YlGnBu", linewidths=.5)  #annotated the heatmap with values
plt.title("Heatmap of a 5x5 Matrix")
plt.show()
```

# 8. Comparison of Matplotlib and Seaborn

##Generate a pair of plots (bar charts or line plots) using the same data in both Matplotlib and Seaborn. Highlight the differences in the code, syntax, and aesthetics of the two libraries.

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]

plt.figure(figsize=(8, 6))  # adjusting the figure size
plt.bar(categories, values, color='skyblue')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Matplotlib Bar Chart")
plt.show()


plt.figure(figsize=(8, 6))  # adjusting the figure size
sns.barplot(x=categories, y=values, palette="Blues_d")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Seaborn Bar Plot")
plt.show()
```

#10. Save and Style Plots

##a) Plot any graph of your choice and save it in both .png and .pdf formats.

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(8, 6))  # adjusting the figure size
plt.scatter(x, y, color='darkblue', marker='o', s=50, alpha=0.7, label='Data Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Scatter Plot')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)


plt.savefig('scatter_plot.png', dpi=300, bbox_inches='tight')
plt.savefig('scatter_plot.pdf', bbox_inches='tight')

plt.show()
```
